```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# There are three datasets provided: amazon, imdb, yelp.
# First one to be analyzed is Amazon

# site of raw Amazon data

dfa = pd.read_csv(r'C:\Users\jafon\Documents\PythonMaterials\Data\sentimentamazon_imdb_yelp\amazon_reviews.txt', 
                  delimiter= '\t', header=None)
dfa.columns = ['review', 'Bad 0 or Good 1']
dfa.head()
```

```python
# Creating a feature - keyword sentiment analysis

# first creating moddable list of words that can be used as flag words for bad
keywordsbad = ['problem', 'fool', 'mislead', 'poor', 'return', 'returned', 'fooled', 'problems',
           'problematic', 'misleading', 'broken', 'broke', 'disappointed', 'bad', 'beware', 'flaw', 'flawed', 
           'failed', 'waste', 'useless', 'worst', 'embarassing', 'junk', 'crap']

keywordsgood = ['good', 'great', 'pleased', 'recommend', 'recommended', 'superb', 'super', 'excellent', 'incredible',
               'best', 'authentic', 'love', 'perfect', 'perfectly', 'happy', 'happier', 'satisfied', 'loved']

```

```python
# I am now using these keywords to create a sentiment model
# I could've used a regex system to avoid unnecessary columns with similar words 
# (e.g., 'return', 'returned', and 'problems', 'problematic'), but I didn't want to overcomplicate the code

for key in keywordsbad:
    # Note that we add spaces around the key so that we're getting the *word*,
    # not just pattern matching.
    dfa[str(key)] = dfa.review.str.contains(
        ' ' + str(key) + ' ',
        case=False # ignoring the case of the keyword
    )
    
for key in keywordsgood:
    dfa[str(key)] = dfa.review.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
dfa.head()
```

```python
# also adding in all caps to be bad
dfa['capslock'] = dfa.review.str.isupper()
dfa.head()
```

```python
dfa.head(20)
```

```python
# checking correlations for appropriateness of a Naive Bayes model

ax = plt.figure(figsize=(15,10))

sns.heatmap(dfa.corr(), cmap="YlGnBu")
plt.show()
```

```python
# It appears with the exception of a couple words (i.e., junk x broke, crap x worst, 
# capslock x junk, capslock x returned), all x factors are independent, making Naive Bayes appropriate
# Not sure what the white lines on the corr heatmap mean - ask mentor!

#------------------------------------------------------------------------------------------

# use bernoulli model, apply to yelp...

from sklearn.naive_bayes import BernoulliNB

# creating an abbreviation 
bnb = BernoulliNB()

# creating lists to populate the model with
data = dfa[keywordsbad + keywordsgood + ['capslock']] # looking at the x factors
target = dfa['Bad 0 or Good 1'] # this summarizes the target sentiment

#apply the model to the data
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results. This syntax is taken from the Thinkful curriculum.
# This code is to determine how many points that the predictor model differs from 
# the actual Bad or Good conclusion - i.e., lower mislabeled points -> higher accuracy
# Also, the instructions dictate using this accuracy measure model

print("Number of mislabeled points out of a total {} points for the Amazon Naive Bayes Model : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```

```python
# ^^^^could be [a lot] better - a threshold accuracy was not sought after from the outset.

#applying the classifiers to the yelp dataset
dfy = pd.read_csv(r'C:\Users\jafon\Documents\PythonMaterials\Data\sentimentamazon_imdb_yelp\yelp_reviews.txt', 
                  delimiter= '\t', header=None)
dfy.columns = ['review', 'Bad 0 or Good 1']
dfy.head(2)
```

```python
# setting up this new dataset for sentiment population

for key in keywordsbad:
    dfy[str(key)] = dfy.review.str.contains(
        ' ' + str(key) + ' ',
        case=False)
    
for key in keywordsgood:
    dfy[str(key)] = dfy.review.str.contains(
        ' ' + str(key) + ' ',
        case=False)

dfy['capslock'] = dfy.review.str.isupper()
dfy.head(4)
```

```python
# applying the model to the new dataset
bnb = BernoulliNB()

# creating lists to populate the model with
data_yelp = dfy[keywordsbad + keywordsgood + ['capslock']] # looking at the x factors
target_yelp = dfy['Bad 0 or Good 1'] # this summarizes the target sentiment

#apply the model to the data
bnb.fit(data_yelp, target_yelp)

# Classify, storing the result in a new variable.
y_pred_yelp = bnb.predict(data_yelp)

# Display our results. This syntax is taken from the Thinkful curriculum.
# This code is to determine how many points that the predictor model differs from 
# the actual Bad or Good conclusion - i.e., lower mislabeled points -> higher accuracy
# Also, the instructions dictate using this accuracy measure model

print("Number of mislabeled points out of a total {} points for the Yelp Naive Bayes Model : {}".format(
    data_yelp.shape[0],
    (target_yelp != y_pred_yelp).sum()
))
```

```python
# breaking down the above via a confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(target_yelp, y_pred_yelp)
```

```python
# read this as:
# [1,2
#  3,4]:


# 1: Model predicted good review - was actually good review   (True Positive)
# 2: Model predicted bad review - but was actually good review (False Negative)
# 3. Model predicted good review - was actually bad review     (False Positive)
# 4. Model predicted bad review - was actually bad review    (True Negative)
```

```python
# testing FIRST ITERATION of naive bayes classifier accuracy with single holdout group

# holdout group is 20% of dataset

from sklearn.model_selection import train_test_split

# Use train_test_split to create the necessary training and test groups

X_train, X_test, y_train, y_test = train_test_split(data_yelp, target_yelp, test_size=0.2, random_state=20)
print('With 20% Holdout: ' + str(bnb.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(bnb.fit(data_yelp, target_yelp).score(data_yelp, target_yelp)))
```
