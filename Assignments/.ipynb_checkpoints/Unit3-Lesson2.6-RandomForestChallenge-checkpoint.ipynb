# Goal:

1. Build decision tree
2. Build random forest
3. Compare runtime of decision tree with runtime of random forest.

```python
# for the record, runtime is calculated by:
import time
start_time = time.time()

print("--- %s seconds ---" % round((time.time() - start_time), 3))
```

```python
# deprecation warning?! Bah, humbug!
# time to set training variables

Xtrain = dftrain[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].copy()
Ytrain = dftrain['Survived']
```

```python
Xtest = dftest[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].copy()
```

```python
# welp, learned my lesson trying to let python do the heavy lifting

#encoding my categorical data

from sklearn.preprocessing import LabelEncoder

labelencoder_X = LabelEncoder()
Xtrain = labelencoder_X.fit_transform(Xtrain)
Xtest = labelencoder_X.transform(Xtest)
```

```python
Xtest.shape
```

```python
# I have had it with these snakes in these planes. I'll just do get_dummies

Xtrain2 = pd.get_dummies(Xtrain)
Xtest2 = pd.get_dummies(Xtest)
```

```python
Xtrainnames = list(Xtrain2.columns)
Xtestnames = list(Xtest2.columns)
```

```python
Xtrain2.head()
```

```python
from sklearn.preprocessing import Imputer
train_imputer = Imputer(missing_values = np.nan, strategy = 'mean', axis = 0)
train_imputer = train_imputer.fit(Xtrain2)
Xtrain2 = train_imputer.transform(Xtrain2)

test_imputer = Imputer(missing_values = np.nan, strategy = 'mean', axis = 0)
test_imputer = test_imputer.fit(Xtest2)
Xtest2 = test_imputer.transform(Xtest2)
```

```python
# The necessary transformations *SHOULD* be complete by now...

# time to build decision tree classifier

from sklearn import tree

# A convenience for displaying visualizations.
from IPython.display import Image

# Packages for rendering our tree.
import pydotplus
import graphviz

# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='gini',
    max_depth=10,
)
decision_tree.fit(Xtrain2, Ytrain)

print("--- %s milliseconds ---" % round((time.time() - start_time), 4)) # for runtime!!!!
```

```python
# Render our tree graphically:

dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=Xtrainnames,
    class_names=['Did not Survive', 'Survive'],
    filled=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
decision_tree.score(Xtrain2, Ytrain)
```

```python
# trying a cross val

from sklearn.model_selection import cross_val_score

print("--- %s milliseconds ---" % round((time.time() - start_time), 4))

cross_val_score(decision_tree, Xtrain2, Ytrain, cv=10)
```

```python
Ypredict = decision_tree.predict(Xtest2)
```

```python
decision_tree.score(Xtest2, Ypredict) # ?????
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4))

cross_val_score(decision_tree, Xtest2, Ypredict, cv=10)
```

```python
# Ah, now I see what they mean by decision trees tend to overfit. May need to limit depth next time...
```

```python
decision_tree2 = tree.DecisionTreeClassifier(
    criterion='gini',
    max_depth=7, # reducing depth to avoid overfitting
    max_features=5 # setting limit to features too to avoid overfitting
)
decision_tree2.fit(Xtrain2, Ytrain)

print("--- %s milliseconds ---" % round((time.time() - start_time), 4)) # for runtime!!!!
```

```python
decision_tree2.score(Xtrain2, Ytrain)
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4))

cross_val_score(decision_tree2, Xtrain2, Ytrain, cv=10)
```

```python
Ypredict2 = decision_tree2.predict(Xtest2)
decision_tree2.score(Xtest2, Ypredict2) 
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4))

cross_val_score(decision_tree2, Xtest2, Ypredict2, cv=10)
```

```python
# Now to jump into the Random Forest Model
from sklearn import ensemble
rfc = ensemble.RandomForestClassifier(max_depth=10)
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4)) # for runtime!!!!
rfc.fit(Xtrain2, Ytrain)
rfc.score(Xtrain2, Ytrain)
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4)) # for runtime!!!!

cross_val_score(rfc, Xtrain2, Ytrain, cv=10)
```

```python
Ypredict3 = rfc.predict(Xtest2)
```

```python
rfc.score(Xtest2, Ypredict3) 
```

```python
print("--- %s milliseconds ---" % round((time.time() - start_time), 4)) # for runtime!!!!

cross_val_score(rfc, Xtest2, Ypredict3, cv=10)
```

It seems that the decision tree is at least 25x faster than the random forest, even though the depth is the same. This means that random forest is much slower, as n_estimators = 10, and so decision tree should only be about 10x faster.
