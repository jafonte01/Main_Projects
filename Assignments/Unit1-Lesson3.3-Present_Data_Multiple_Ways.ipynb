```python
dflimit = df[df.price < 2000]
```

```python
dflimit.head()
```

```python
plt.hist(x=dflimit.price, bins=50)
plt.title("Distribution of Amsterdamn AirBnB Prices under 2000 USD")
plt.xlabel("Dollars (USD)")
plt.show()
```

```python
sns.kdeplot(data=dflimit.price, shade=True, legend=False)
plt.title("Kernel Density Distribution of Prices for Amsterdam AirBnB")
plt.xlabel("Price (USD)")
plt.ylabel("Relative Density Distribution")
plt.show()
```

```python
#do pie chart of intervals
#do scatter
plt.scatter(x=df.price, y=df.price)
plt.title("The World's Most Unhelpful Single-Variable Graph")
plt.xlabel("Price of Amsterdam AirBnB (in USD)")
plt.ylabel("Price of Amsterdam AirBnB (in USD)")
plt.show()
```

```python
#creating intervals
df0to100 = df[df.price <=100]
df100to200 = df[(df.price > 100) & (df.price <= 200)]
df200to300 = df[(df.price > 200) & (df.price <= 300)]
df300to400 = df[(df.price > 300) & (df.price <=400)]
df400to500 = df[(df.price > 400) & (df.price <=500)]
df500plus =  df[df.price > 500]

```

```python
#Creating pie graph
# the "-1" is used to not account for the column name line

pieprice = [df0to100.price.count()-1, df100to200.price.count() - 1, df200to300.price.count() - 1, df300to400.price.count() - 1,
            df400to500.price.count() - 1, df500plus.price.count() - 1]
piepricelabels = ['\$100 and less', '\$100 to \$200', '\$200 to \$300', '\$300 to \$400', '\$400 to \$500', '\$More than $500']

plt.figure(figsize=(15,5)) 
plt.pie(pieprice, autopct='%d%%', labels=piepricelabels)
plt.axis('equal')
plt.title('Breakdown of Nightly AirBnB Prices in Amsterdam (in USD)')
plt.show()
```

```python
#2 - Bivariate plotting - three different ways
#kde
#scatter / lmplot
# ????
dfkde = dflimit[["price", "minimum_nights"]].copy()
sns.kdeplot(data=dfkde, shade=True, legend=False, style='dark grid')
plt.title("Bivariate Kernel Density Distribution of Prices and Minimum Nights for Amsterdam AirBnB")
plt.xlim(0,300)
plt.ylim(0,20)
plt.xlabel("Price (USD)")
plt.ylabel("Minimum Nights to be Reserved")
plt.show()
```

```python
plt.scatter(x="minimum_nights", y="price", data=dflimit)
plt.title("Price of Amsterdam AirBnB over Minimum Nights required to Reserve")
plt.xlabel("Minimum Nights")
plt.xlim(-10,400)
plt.ylabel("Price (USD)")
plt.show()
```

```python
dfpriceavg = [df0to100.price.mean(), df100to200.price.mean(), df200to300.price.mean(), 
              df300to400.price.mean(), df400to500.price.mean(), df500plus.price.mean()]
dfnightsavg = [df0to100.minimum_nights.mean(), df100to200.minimum_nights.mean(), df200to300.minimum_nights.mean(), 
              df300to400.minimum_nights.mean(), df400to500.minimum_nights.mean(), df500plus.minimum_nights.mean()]
```

```python
dfprivate = dflimit[df.room_type == "Private room"]
dfnotprivate = dflimit[df.room_type != "Private room"]
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(3,1,1)
ax = sns.boxplot(x='price',data=dfprivate)  
plt.title('Price Distribution for Private Rooms')
plt.xlim(0,1000)


plt.subplot(3,1,3)
ax = sns.boxplot(x='price', data=dfnotprivate)
plt.title('Price Distribution for Non-Private Rooms')
plt.xlim(0,1000)

plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(3,1,1)
ax = sns.barplot(x='price',data=dfprivate)  
plt.title('Price Distribution for Private Rooms')
plt.xlim(0,200)


plt.subplot(3,1,3)
ax = sns.barplot(x='price', data=dfnotprivate)
plt.title('Price Distribution for Non-Private Rooms')
plt.xlim(0,200)

plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(3,1,1)
ax = sns.violinplot(x='price',data=dfprivate)  
plt.title('Price Distribution for Private Rooms')
plt.xlim(0,500)


plt.subplot(3,1,3)
ax = sns.violinplot(x='price', data=dfnotprivate)
plt.title('Price Distribution for Non-Private Rooms')
plt.xlim(0,500)

plt.show()
```

```python
plt.hist(x=dfprivate.price, alpha=0.3, color='orange', bins=40, label='Private Rooms')
plt.hist(x=dfnotprivate.price, alpha=0.3, color='blue', bins=40, label='Non-Private Rooms')
plt.title('Distributions of Price for Private and Non-Private Rooms')
plt.xlabel('Price (USD)')
plt.xlim(0,750)
plt.legend()
plt.show()
```

```python
p1=sns.kdeplot(dfprivate.price, shade=True, color="r", label='Private Rooms')
p1=sns.kdeplot(dfnotprivate.price, shade=True, color="b", label='Non-Private Rooms')
plt.title('Density Distribution of Price (in USD) for Private vs. Non-Private AirBnB Rooms')
plt.xlabel('Price')
plt.xlim(0,500)
plt.show()
```

```python

b = sns.factorplot(y="price", data=dfprivate,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=True)

b.set_ylabels("Price")
b.set_xlabels("Private Rooms")
plt.title('Pointplot - Price by Categorical Variable Private Room or Non-Private AirBnB Room')


c = sns.factorplot(y="price", data=dfnotprivate,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=True, color='orange')
c.set_xlabels("Non-Private Rooms")

plt.show()
```
